### Problema 1

```python
# Problema 1: Clasificador de temperatura

# Utilicé un ciclo 'for' con 'range' para recorrer los números desde -30 hasta 50.
# Nota: Ponemos 51 porque la función range() no incluye el último número.
for temperatura in range(-30, 51):

    # Condición 1: Si el número es menor que 0
    if temperatura < 0:
        print(f"{temperatura} Muy frío")

    # Condición 2: Si el número está entre 0 y 24.
    # Como ya descartamos los menores a 0 en el 'if' anterior,
    # solo necesitamos comprobar que sea menor que 25.
    elif temperatura < 25:
        print(f"{temperatura} Templado")

    # Condición 3: Si el número es mayor o igual a 25.
    # Si no fue ni menor a 0 ni menor a 25, por descarte es 25 o mayor.
    else:
        print(f"{temperatura} Caluroso")
```

### Problema 2

```python
# Problema 2: Números especiales del robot

# Utilicé un ciclo 'for' para recorrer los números del 1 al 100.
# Puse 101 para que el 100 esté incluido en el ciclo.
for numero in range(1, 101):

    # IMPORTANTE: Primero verificar si es divisible por 10.
    # Si un número es divisible por 10, automáticamente también es divisible por 5.
    # Por eso debemos atrapar los múltiplos de 10 antes que los de 5.
    if numero % 10 == 0:
        print(f"{numero} Bip Bip")

    # Si no fue divisible por 10, ahora sí verificamos si es divisible por 5.
    elif numero % 5 == 0:
        print(f"{numero} Bip")

    # Si el número no es divisible por 10 ni por 5, simplemente no hacemos nada
    # y el ciclo continúa con el siguiente número, tal como pide el problema.
```

### Problema 3



```python
# Problema 3: Detector de puntaje gamer

# Solicitamos al usuario que ingrese los dos puntajes.
# Usamos input() para pedir el dato y int() para convertir ese texto en un número entero.
puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

# Sumamos ambos puntajes y guardamos el resultado en una nueva variable.
suma = puntaje1 + puntaje2

# Evaluamos el resultado de la suma según las reglas dadas:

# Regla 1: si la suma es menor que 50
if suma < 50:
    print("\nResultado: Nivel principiante")

# Regla 2: si la suma es mayor o igual a 50 y menor que 80
# Como el 'if' anterior ya filtró todos los números menores a 50,
# si el código llega aquí, ya sabemos que es 50 o mayor.
# Por lo tanto, solo necesitamos comprobar si es menor que 80.
elif suma < 80:
    print("\nResultado: Nivel intermedio")

# Regla 3: si la suma es mayor o igual a 80
# Si la suma no fue menor a 50, ni tampoco menor a 80, por descarte es 80 o más.
else:
    print("\nResultado: Nivel pro")
```

### Problema 3


```python
# Problema 4: Calculador de perfil musical (Versión interactiva)

# 1. Solicitamos la edad y la convertimos a número entero
edad = int(input("Ingresa tu edad: "))

# 2. Preguntamos por la música y traducimos su respuesta
respuesta_musica = input("¿Te gusta la música? (responde 'si' o 'no'): ")

# Si el usuario escribió "si", asignamos True a la variable requerida
if respuesta_musica == "si":
    le_gusta_musica = True
else:
    le_gusta_musica = False

# 3. Preguntamos por el baile y traducimos su respuesta
respuesta_bailar = input("¿Te gusta bailar? (responde 'si' o 'no'): ")

if respuesta_bailar == "si":
    le_gusta_bailar = True
else:
    le_gusta_bailar = False

# 4. Evaluamos el perfil usando las tres variables

# Regla 1: Tiene 18 o más años Y le gusta la música Y le gusta bailar.
if edad >= 18 and le_gusta_musica == True and le_gusta_bailar == True:
    print("\nEres una estrella de la fiesta")

# Regla 2: Le gusta la música Y NO le gusta bailar.
elif le_gusta_musica == True and le_gusta_bailar == False:
    print("\nTe gusta la música, pero no tanto bailar")

# Regla 3: Cualquier otro caso en el que no se cumpla lo anterior.
else:
    print("\nTienes tu propio estilo")
```
